# 1.1 Fit Evaluation Criteria

## 1.1.1 Chi-squared (chi2)

The chi-squared test is a statistical test used to determine whether there is a significant difference between the expected frequencies and the observed frequencies in one or more categories. It is commonly used to test hypotheses about categorical data, such as whether certain attributes or characteristics are related to a particular outcome.

The formula for the chi-squared test statistic is:

$$
χ^{2} = \sum\frac{(O - E)^{2}}{E}
$$

Where:

* $χ^{2}$ is the chi-squared test statistic<br>
* $O$ is the observed frequency in each category<br>
* $E$ is the expected frequency in each category, calculated as (row total * column total) / grand total

The degrees of freedom for the chi-squared test is $(r-1)(c-1)$ where $r$ is the number of rows and $c$ is the number of columns in the contingency table.

Once you have calculated the chi-squared test statistic, you can compare it to a chi-squared distribution with the appropriate degrees of freedom to determine the p-value. If the $p\text{* -}value$ is less than your chosen significance level (typically 0.05), you can reject the null hypothesis and conclude that there is a significant difference between the observed and expected frequencies.

## 1.1.2 Akaike Information Criterion (AIC)

The Akaike Information Criterion (AIC) is a measure of the relative quality of a statistical model, used to compare different models and select the best one. It was developed by Hirotugu Akaike in the 1970s. AIC balances the goodness of fit of the model with the complexity of the model, thus it helps to avoid overfitting.

The formula for AIC is:

$$
AIC = 2k - 2ln(L)
$$

Where:

* $k$ is the number of parameters in the model<br>
* $L$ is the likelihood of the data given the model

The smaller the AIC value, the better the model is.

When comparing two models, the model with the lower AIC value is considered to be better, since it has a lower penalty for complexity than the model with the higher AIC value. AIC is a relative measure, meaning that a model is only considered to be better than another model if the difference in their AIC values is greater than 2.

AIC is widely used in model selection, but there are other similar measures like the Bayesian Information Criterion (BIC) which also penalizes the number of parameters, and the sample size adjusted Bayesian Information Criterion (SBC) which takes into account the sample size of the data.

## 1.1.3 Bayesian Information Criterion (BIC)

The Bayesian Information Criterion (BIC) is a measure of the relative quality of a statistical model, similar to the Akaike Information Criterion (AIC). It is used to compare different models and select the best one. BIC was developed by Gideon Schwarz in 1978 and is based on Bayesian statistics.

The formula for BIC is:

$$
BIC = k\;ln(n) - 2ln(L)
$$

Where:

* $k$ is the number of parameters in the model<br>
* $n$ is the sample size<br>
* $L$ is the likelihood of the data given the model

The BIC value is a measure of the trade-off between the goodness of fit of the model and the complexity of the model. A lower BIC value indicates a better model.

When comparing two models, the model with the lower BIC value is considered to be better, since it has a lower penalty for complexity than the model with the higher BIC value. The BIC is more stringent than AIC, meaning that it assigns a higher penalty for complexity, thus it is less likely to select models with a large number of parameters.

BIC is widely used in model selection, it's particularly useful for model selection when sample size is large or moderately large. It's also a powerful tool for variable selection in multiple regression analysis.

## 1.1.4 Mean Absolute Error (MAE)

Mean Absolute Error (MAE) is a measure of the average magnitude of the errors in a forecast, without considering their direction. It measures the average of the absolute differences between the forecasted and actual values.

The formula for Mean Absolute Error is:

$$
MAE = (\frac{1}{n}) \sum|y - ŷ|
$$

Where:

* $y$ is the actual value<br>
* $ŷ$ is the predicted or forecasted value<br>
* $n$ is the number of observations

The MAE is a linear score which means that all the individual differences are weighted equally in the average. It is a commonly used measure of accuracy for continuous variables and it is easy to understand and interpret.

MAE is also scale-dependent, meaning that the error will change depending on the scale of the variables. It's often used alongside with Mean Squared Error (MSE) and Root Mean Squared Error (RMSE) which are also scale-dependent measures of accuracy.

It's widely used in regression and time series analysis to evaluate the performance of the model, and also in other fields like image processing or natural language processing.

## 1.1.5 Mean Square Error (MSE)

Mean Square Error (MSE) is a measure of the difference between the forecasted and actual values. It is a commonly used metric for evaluating the performance of a model, particularly in the context of regression and time series analysis. MSE is the mean of the squared differences between predicted and actual values.

The formula for Mean Square Error is:

$$
MSE = (\frac{1}{n}) \sum(y - ŷ)^{2}
$$

Where:

* $y$ is the actual value<br>
* $ŷ$ is the predicted or forecasted value<br>
* $n$ is the number of observations

MSE is a measure of the average of the squared differences between the predicted and actual values and it is in squared units. It is sensitive to outliers as it squares the residuals before averaging, thus large errors will have a relatively larger effect on the final value.

As it squares the errors, it gives more weight to larger errors, and it is widely used in regression and time series analysis to evaluate the performance of the model. It is also often used alongside with Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE), which are also commonly used measures of accuracy.

## 1.1.6 Root Mean Square Error (RMSE)

Root Mean Square Error (RMSE) is a measure of the difference between the forecasted and actual values. It is a commonly used metric for evaluating the performance of a model, particularly in the context of regression and time series analysis. RMSE is the square root of the mean of the squared differences between predicted and actual values.

The formula for Root Mean Square Error is:

$$
RMSE = \sqrt{(\frac{1}{n})\sum(y - ŷ)^{2}}
$$

Where:

* $y$ is the actual value<br>
* $ŷ$ is the predicted or forecasted value<br>
* $n$ is the number of observations

RMSE is a measure of the absolute differences between the predicted and actual values and it is in the same unit as the variable being predicted. It is sensitive to outliers as it squares the residuals before averaging, thus large errors will have a relatively larger effect on the final value.

As it squares the errors, it gives more weight to larger errors, and it is widely used in regression and time series analysis to evaluate the performance of the model. It is also often used alongside with Mean Absolute Error (MAE) and Mean Squared Error (MSE), which are also commonly used measures of accuracy.

## 1.1.7 Root Mean Square Error of Approximation (RMSEA)

The Root Mean Square Error of Approximation (RMSEA) is a measure of fit for structural equation models (SEM) that provides a way to evaluate the accuracy of the model in explaining the covariance structure of the data. It is a scalar value between 0 and 1, with lower values indicating a better model fit.

The formula for RMSEA is:

$$
RMSEA = \frac{\sqrt{\frac{SSE}{df}}}{\frac{SST}{n}}
$$

Where:

* $SSE$ is the residual sum of squares<br>
* $df$ is the degrees of freedom of the residuals<br>
* $SST$ is the total sum of squares<br>
* $n$ is the sample size<br>

RMSEA is a relative measure of fit and it is sensitive to sample size. Therefore, it is recommended to use the 90% Confidence Interval (C.I) for RMSEA as well. When the C.I for RMSEA doesn't include 0 and .08, it is considered as an acceptable fit.

An RMSEA value of 0 indicates a perfect fit, while a value greater than 0 indicates that there is some discrepancy between the observed covariance matrix and the estimated one. Generally, values less than 0.05 are considered to indicate a good fit, while values greater than 0.10 are considered to indicate a poor fit.

## 1.1.8 Comparative Fit Index (CFI)

The Comparative Fit Index (CFI) is a measure of model fit that compares the fit of a proposed model to a baseline model, usually a null model that represents the minimum possible fit. It is widely used in structural equation modeling (SEM) and other forms of factor analysis.

CFI is calculated by the following formula:

$$
CFI = \frac{S-Snull}{S-Snull + pnull-p}
$$

Where:

* $S$ is the model's goodness of fit statistic<br>
* $Snull$ is the goodness of fit statistic for the null model<br>
* $p$ is the number of parameters in the model<br>
* $pnull$ is the number of parameters in the null model

A CFI value of 1 indicates a perfect fit, while a value less than 1 indicates a worse fit than the null model. A value greater than 0.95 is generally considered as an acceptable fit, but values between 0.90 and 0.95 are considered marginal.

It is important to note that, CFI should be used in conjunction with other fit indices, such as RMSEA and GFI, to get a comprehensive evaluation of the model's fit.

## 1.1.9 Goodness of Fit Index (GFI)

The Goodness-of-Fit Index (GFI) is a measure of the fit of a proposed model to the observed data. It is widely used in structural equation modeling (SEM) and other forms of factor analysis. The GFI is a statistic that ranges from 0 to 1, with a value closer to 1 indicating a better fit between the observed data and the model.

The formula for GFI is:

$$
GFI = \frac{Snull - S}{Snull}
$$

Where:

* $S$ is the model's goodness of fit statistic<br>
* $Snull$ is the goodness of fit statistic for the null model (i.e., the model with no independent variables)

A GFI value of 1 indicates a perfect fit, while a value less than 1 indicates a worse fit than the null model. A value greater than 0.90 is generally considered to indicate an acceptable fit, although values between 0.80 and 0.90 are sometimes considered to be marginal.

GFI should be used along with other fit indices like the Root Mean Square Error of Approximation (RMSEA) and the Comparative Fit Index (CFI) to get a comprehensive assessment of the model's fit. It should be noted that, GFI is sensitive to sample size and it is recommended to use it along with other indices that are insensitive to sample size.

## 1.1.10 Adjusted R-squared

The adjusted R-squared is a modified version of the R-squared statistic that adjusts for the number of predictor variables in a model. R-squared is a commonly used measure of the goodness of fit of a linear regression model, it ranges from 0 to 1, and its value closer to 1 indicates a better fit of the model to the data. However, R-squared increases as the number of predictor variables in the model increases, even if these predictor variables do not improve the model's ability to predict the response variable.

The formula for the adjusted R-squared is:

$$
\text{Adjusted R-squared} = 1 - (\;(1-R²) * \frac{n-1}{n-p-1}\;)
$$

Where:

* $R^{2}$ is the regular R-squared value<br>
* $n$ is the sample size<br>
* $p$ is the number of predictor variables (not including the constant term)

The adjusted R-squared value ranges from 0 to 1, similar to R-squared and a value closer to 1 indicates a better fit of the model to the data. However, the adjusted R-squared penalizes the model for having a large number of predictor variables, unlike the R-squared.

Adjusted R-squared is a useful measure for comparing models with different numbers of predictor variables, it can help to determine whether the increase in R-squared from adding more variables is due to the variables themselves or just due to the increased number of variables in the model.

## 1.3 Metrics Implementation 
Implementing the metrics on the adult income dataset <br>
we have : <br>
**The feature selection metrics:**
- Chi-squared test
- Mutual Information
- Recursive Feature Elimination (RFE) <br>

**The model evaluation metrics :**
- AIC
- BIC
- MAE
- MSE
- RMSE
- CFI
- GFI 
- R-Squared <br>

We are going to use the sklearn library to calculate these Metrics 

```python
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Load the dataset
data = dataset
X = data.drop("income", axis=1)
y = data["income"]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#We decided to split the dataset into test and train sets without a validation set as a way to quickly implement the process.


## Feature Selection

# Chi-Squared test
chi_selector = SelectKBest(score_func=chi2, k=5)
chi_selector.fit(X_train, y_train)
chi_selected_features = X.columns[chi_selector.get_support()]

# Mutual Information
mi_selector = SelectKBest(score_func=mutual_info_classif, k=5)
mi_selector.fit(X_train, y_train)
mi_selected_features = X.columns[mi_selector.get_support()]

# Recursive Feature Elimination
rfe_selector = RFE(estimator=LogisticRegression(), n_features_to_select=5)
rfe_selector.fit(X_train, y_train)
rfe_selected_features = X.columns[rfe_selector.support_]

# Apply the selected features on the model
X_train_fs = X_train[chi_selected_features]
X_test_fs = X_test[chi_selected_features]

## Model Evaluation

# Train the model
log_reg = LogisticRegression(solver="saga")

log_reg.fit(X_train_fs, y_train)

# Make predictions on the test set
y_pred = log_reg.predict(X_test_fs)

# Calculate the mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_pred)
# absolute_error(y_test, y_pred)

# Calculate the mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the root mean squared error (RMSE)
rmse = np.sqrt(mse)

# Calculate the coefficient of determination (R-squared)
r2 = r2_score(y_test, y_pred)

# Print the results
print("Selected features (Chi-Squared):", chi_selected_features)
print()
print("Selected features (Mutual Information):", mi_selected_features)
print()
print("Selected features (RFE):", rfe_selected_features)
print()
print("Mean Absolute Error (MAE):", mae)
print()
print("Mean Squared Error (MSE):", mse)
print()
print("Root Mean Squared Error (RMSE):", rmse)
print()
print("R-squared:", r2)
```

### Cost Function
A Cost Function/Loss Function evaluates the performance of our Machine Learning Algorithm. The Loss function computes the error for a single training example, while the Cost function is the average of the loss functions for all the training examples.The results are then added together and we get the most commonly used cost function, mean squared error (MSE).
$$
MSE =\frac{1}{N}\sum_{n=1}^{N} (Y'-Y)^2
$$
Our aim here is to find those values of m and b iny = mx+b For which the error is minimum.Values that minimize the cost function <br>
**values that minimize the cost function**
### How do we minimize any function?
If we look carefully, our Cost function is of the form $Y = X^2$ . In a Cartesian coordinate system, this is an equation for a parabola and can be graphically represented as : 
To minimize the function above, we need to find that value of X that produces the lowest value of Y which is the red dot. It is pretty easy to locate the minima here since it is a 2D graph, but this may not always be the case, especially in higher dimensions. For those cases, we need to devise an algorithm to locate the minima, and that algorithm is called **Gradient Descent**.

# 2.2 Conjugate Gradient Algorithms

## 2.2.1 Introduction

The conjugate gradient $(\mathrm{CG})$ method is one of the most popular and well known iterative techniques for solving sparse symmetric positive definite(SPD) systems of linear equations. It was originally developed as a direct method, but became popular for its properties as an iterative method, especially following the development of sophisticated preconditioning techniques.

The intention behind this small note is to present the motivation behind the CG-method, followed by some details regarding the algorithm and its implementation. It is important to note that the mathematical derivation of the method is intentionally incomplete and missing important details. This is done in order to keep the note small and focus on the more intuitive characteristics of the CG-method, making it more suitable for a target audience with a limited mathematical background.

## 2.2.2 The Quadratic Test Function

A natural starting point in deriving the conjugate gradient method is by looking at the minimization of the quadratic test function

$$
\phi(x)=\frac{1}{2} x^{T} A x-x^{T} b
$$

with $b, x \in \mathbb{R}^{n}$ and $A \in \mathbb{R}^{n x n}$, where $A$ is assumed to be SPD.

Note: A symmetric matrix $A$ is $S P D$ (symmetric positive definite) if $x^{T} A x>$ $0 \forall x \in \Omega$, or equivalently if all the egeinvalues of $A$ are positive.

The minimizer $x^{*}$ of the function $\phi$ is given as the point where the gradient of the function is equal to zero. Direct calculation gives

$$
\nabla \phi\left(x^{*}\right)=A x^{*}-b=0
$$

or

$$
A x^{*}=b
$$

This shows that the same $x^{*}$ that minimizes $\phi(x)$ also serves as the solution to the linear system of equations $A x=b$, motivating us to find a method capable of minimizing $\phi$ instead of looking at a direct solver of the algebraic system.

The uniqueness of our solution is guaranteed by the SPD condition. To see this, consider what you learned about unqiue minimas and maximas in high school(vgs). A minimum point is unique and global if the double derivative is larger than zero for all points of the domain. For multiple dimensions this is the same as requiring that the hessian(or hess matrix) $H=\nabla^{2}>0$. For the quadratic test function this means that $\nabla^{2} \phi(x)=A$. But since $A$ is SPD,$x^{T} A x>0$, and hence our solution $x^{*}$ is unique. This uniqueness can also be seen intuitively by considering the strict convexity of the function.


## 2.2.3 Line Search Methods

In the previous section we concluded that minimizing the quadratic test function is equivalent to solving $A x=b$. We now need a strategy for solving our optimization problem.

The line search methods are a large family of iterative optimization methods where the the iteration is given by

$$
x_{k+1}=x_{k}+\alpha_{k} p_{k}
$$

The idea is to choose an initial position $x_{0}$, and for each step walk along a direction(a line) so that $\phi\left(x_{k+1}\right)<\phi\left(x_{k}\right)$. The different methods have various strategies for how they choose the search direction $p_{k}$, and the step length $\alpha_{k}$.

The steepest descent method is perhaps the most intuitive and basic line search method. We remember that the gradient of a function is a vector giving the direction along which the function increases the most. The method of steepest descent is based on the strategy that in any given point $\mathrm{x}$, the search direction given by the negative gradient of the function $\phi(x)$ is the direction of steepest descent. In other words, the negative gradient direction is the locally optimal search direction. We have already found the gradient of $\phi(x)$ to be $A x-b$. We also call this the residual $\mathrm{r}$ of the system.

We now have our direction, but we still need to know how far to walk along it. The natural choice is obviously to walk until we no longer descend, and an expression for the optimal step length $\alpha_{k}$ is easily found to be

$$
\alpha_{k}=\frac{\nabla \phi\left(x_{k}\right)^{T} \nabla \phi\left(x_{k}\right)}{\nabla \phi\left(x_{k}\right)^{T} A \nabla \phi\left(x_{k}\right)}=\frac{r_{k}^{T} r_{k}}{r_{k}^{T} A r_{k}}
$$

by inserting the expression for the next step $x_{k+1}=x_{k}-\alpha \nabla \phi\left(x_{k}\right)$ into the quadratic test function and minimizing with respect to $\alpha$. We repeat this for every step, taking the gradient of $\phi$ in the next point $x_{k+1}$ and by finding a new step length. It is obvious that this leads to search directions that are orthogonal to each other, as can be seen in figure [1]

This means that the method zigzag towards the solution. While it can be shown that the steepest descent method always converge for a SPD problem like ours, it seems obvious that its zigzag behaviour is not the optimal

![caption](ReportFigures/Fig1.png)

and fastest path towards the minimum. The problem is simply that each succesive step is not different enough from the others. We need a method that somehow uses the information from the previous steps in order to avoid running back and forth across the valley of our contour plots.


## 2.2.4 The Conjugate Gradient Method

We now turn our eye to the A-conjugate direction methods. A-conjugacy means that that a set of nonzero vectors $\left\{p_{0}, p_{1}, \ldots, p_{n-1}\right\}$ are conjugate with respect to the SPD matrix $A$. That is

$$
p_{i}^{T} A p_{j} \quad \forall i \neq j
$$

A set of $\mathrm{n}$ such vectors are linearly independent and hence span the whole space $\mathbb{R}^{n}$. The reason why such A-conjugate sets are important is that we can minimize our quadratic function $\phi$ in $\mathrm{n}$ steps by successively minimizing it along each of the directions. Since the set of A-conjugate vectors acts as a basis for $\mathbb{R}^{n}$, we can express the difference between the exact solution $x^{*}$ and our first guess $x_{0}$ as a linear combination of the conjugate vectors:

$$
x^{*}-x_{0}=\sigma_{0} p_{0}+\sigma_{1} p_{1}+\ldots+\sigma_{n-1} p_{n-1}
$$

By utilizing our conjugacy property, we find that the coefficients $\sigma_{k}$ are the same as the step lengths $\alpha_{k}$ that minimize the quadratic function $\phi$ along $x_{k}+\alpha_{k} p_{k}$, and hence:

$$
x^{*}=x_{0}+\alpha_{0} p_{0}+\alpha_{1} p_{1}+\ldots+\alpha_{n-1} p_{n-1}
$$

You can think of this as gradually building the solution along the dimensions of our solution space. In fact, for diagonal matrices, the conjugate search vectors coincide with the coordinate axes. In each step $\mathrm{k}, x_{k}$ is the exact solution $x^{*}$ projected into the solution space spanned by the $\mathrm{k}$ vectors.

All of this sounds great, but so far we have just assumed that the set of A-conjugate search directions exists. In practice we need a way to create it. There are several ways to choose such a set. The eigenvectors of A form a A-conjugate set, but finding the eigenvectors is a task requiring a lot of computations, so we better find another strategy. A second alternative is to modify the usual Gram-Schmidt orthogonalization process. This is also not optimal, as it requires storing all the directions.

It turns out there is a conjugate direction method with the very nice property that each new conjugate vector $p_{k}$ can be computed by using only the previous vector $p_{k-1}$. Without knowing the other previous vectors the new vector is still automatically conjugate to the others. The method works its magic by chosing each new direction $p_{k}$ as a linear combination of the negative residual $-r_{k}$ and the previous search vector $p_{k-1}$. We remember that for the quadratic function the negative residual is equal to the steepest descent or negative gradient direction. Hence the name of the method: the conjugate gradient method.

The formula for the new step becomes

$$
p_{k}=-r_{k}+\beta_{k} p_{k-1}
$$

where $\beta_{k}$ is found by imposing the condition that $p_{k-1}^{T} A p_{k}$ and is given as

$$
\beta_{k}=\frac{r_{k}^{T} A p_{k-1}}{p_{k-1}^{T} A p_{k-1}}=\frac{r_{k}^{T} r_{k}}{r_{k-1}^{T} r_{k-1}}
$$

## 2.2.5 Algorithm

We are finally ready to write up the algorithm for the conjugate gradient method. While we have not covered all the details of its derivation, we should still be able to recognize most of the steps of the iteration, and what they do.

![caption](ReportFigures/Fig2.png)

This is the contour plot of a function, with the steps of the steepest descent method in red and of the conjugate gradient method in green.

<br>

>**The conjugate gradient algorithm**
>
>Compute $r_{0}=A x_{0}-b, p_{0}=-r_{0}$
>
>For $k=0,1,2, .$. until convergence
>
>$$
\begin{aligned}
& \alpha_{k}=\frac{r_{k} T r_{k}}{p_{k}{ }^{T} A p_{k}} \\
& x_{k+1}=x_{k}+\alpha_{k} p_{k} \\
& r_{k+1}=r_{k}+\alpha_{k} A p_{k} \\
& \beta_{k}=\frac{r_{k+1}^{T} r_{k+1}}{r_{k}^{T} r_{k}} \\
& p_{k+1}=-r_{k+1}+\beta_{k} p_{k}
\end{aligned}
$$
>
>**End**

The first step is to find the initial residual which is the same as the first gradient search direction. If the initial solution $x_{0}$ is zero, then $r_{0}$ and $p_{0}$ simply becomes $b$. Inside the for-loop we recognize the first line as the calculation of the step length. In the second line we update our solution by adding our step. Then in line three we update our residual, and line 4 and 5 gives us the new search direction.

For an actual implementation of the algorithm above, there are several things to consider. First we notice that both $r_{k}^{T} r_{k}$ and $A p_{k}$ are each used twice for every iteration. Also $r_{k+1}^{T} r_{k+1}$ is simply $r_{k}^{T} r_{k}$ in the next iteration. There is no sense in doing the same calculations twice, so one should store the results the first time they are calculated and reuse them in the calculations that follow.

Another thing to consider is the matrix-vector calculation $A p_{k}$. When doing a numerical analysis of partial differential equations, our matrix typically comes from a finite difference discretization or from the stiffness assembly of the finite element method. This means that $A$ can be very large, but also very sparse(few non-zero elements) and often with a clear structure(at least for the finite difference methods) such as the tri-diagonal banded matrix from the $2 \mathrm{D}$ poisson problem.

A trick is to avoid construction and storing of the actual matrix, and instead do the equivalent calculations explicitly for each node or element in the problem. This way we can save both valuable memory and floating point operations. Using the specific knowledge of the problem you are trying to solve can often be a key factor in achieving good performance, and in general, numerical solvers that do this will outperform more general solvers.

Finally the 'until convergence' condition in our for-loop is not very precise. As mentioned, CG will theoretically reach the exact solution in no more than $n$ steps where $n$ is the size of the matrix $A$. However, there is no point in doing all those iterations if our solution is already within a tolerable margin of error. Furthermore, when doing numerical analysis of PDEs it is important to remember that the exact solution given by the CG-method is only the exact solution of our discretized system, not the exact solution of our PDE. This means that there is no point in running the CG-method towards an error of value zero(tehcnically this is not even possible, due to the round-off error of floating point numbers), instead one should try to balance the discretization error and the error in the CG-solver.

```python
class LinearClassifier:
    def __init__(self):
        self.weights = None
        
    def fit(self, X, y):
        self.X = X
        self.y = y
        self.x_0 = np.zeros(X.shape[1])
        self.weights = self.conjugate_gradient(self.X, self.y, self.x_0)
    
    def predict(self, X):
        y_pred = np.dot(X, self.weights)
        y_pred[y_pred >= 0] = 1
        y_pred[y_pred < 0] = 0
        return y_pred

    def conjugate_gradient(self, X, y, x_0):
        grad_x = self.grad(X, y, x_0)
        search_direction = -grad_x
        i = 0
        while i < 10000:
            alpha = -np.dot(grad_x, search_direction) / np.dot(search_direction, search_direction)
            x_new = x_0 + alpha * search_direction
            grad_x_new = self.grad(X, y, x_new)
            beta = np.dot(grad_x_new, grad_x_new) / np.dot(grad_x, grad_x)
            search_direction = -grad_x_new + beta * search_direction
            if np.linalg.norm(x_new - x_0) < 1e-5:
                break
            x_0 = x_new
            grad_x = grad_x_new
            i += 1
        return x_0
    
    def obj_fun(self, X, y, weights):
        y_pred = np.dot(X, weights)
        y_pred[y_pred >= 0] = 1
        y_pred[y_pred < 0] = 0
        return -np.mean(y == y_pred)
    
    def grad(self, X, y, weights):
        y_pred = np.dot(X, weights)
        y_pred[y_pred >= 0] = 1
        y_pred[y_pred < 0] = 0
        return -np.dot(X.T, y_pred - y) / len(y)
```

## 2.3.2 Grid Search

Grid searching is a method to find the best possible combination of hyper-parameters at which the model achieves the highest accuracy.<br>
The goal of grid search is to find the set of hyperparameters that produce the best performance for a given model. This is done by specifying a set of possible values for each hyperparameter, and then training and evaluating the model for all possible combinations of those values.

### 2.3.2.1 Grid Search Algorithm

### Mathematically
Mathematically, it can be represented as a search through a hyperparameter space H, where each element $h = (h_1, h_2, …h_m)$ in H corresponds to a set of hyperparameter values.Given a model $f(x,h)$ where $x$ is the input data and $h$ is the set of hyperparameters, the goal of grid search is to find the set of hyperparameters h* that minimizes the loss function L(f(x,h),y) where y is the output data.The best set of parameter values is then selected based on the performance metric.

### How Grid Search Works

Grid Searching is a method of optimizing the performance of a model by tuning its hyperparameters. It can be applied to any algorithm that can benefit from hyperparameter tuning, such as **K-Nearest Neighbors** and **Logistic Regression**.<br><br>
For example, in the case of K-Nearest Neighbors, we can use grid search to find the optimal value of K by evaluating the performance of the model for different values of K. Similarly, for Logistic Regression, we can use grid search to find the best learning rate by evaluating the performance of the model for different learning rates.
<br><br>
In both cases, Grid Searching allows us to systematically explore the hyperparameter space and find the set of hyperparameters that lead to the best performance of the model.

## 2.3.3 Random Search

The most efficient way to find an optimal set of hyperparameters for a machine learning model is to use random search.<br>
It is an alternative to grid search, which tries all possible combinations of a predefined set of hyperparameter values. Instead, random search samples the hyperparameter space randomly, and returns the best set of parameters based on a performance metric.<br>

The randomized search meta-estimator is an algorithm that trains and evaluates a series of models by taking random draws from a predetermined set of hyperparameter distributions. The algorithm picks the most successful version of the model it’s seen after training N different versions of the model with different randomly selected hyperparameter combinations, leaving you with a model trained on a near-optimal set of hyperparameters. Note that random search is called a meta-estimator because it’s not an estimator in itself, but rather an algorithm applied to an existing estimator in order to tune the estimator’s hyperparameters.

### 2.3.3.1 Random Search algorithm 

Random local search (or random search) is a optimization algorithm used to find the minimum of a function. It is considered a local optimization method, which means that it only explores the neighborhood of the current point (the point at the current iteration) rather than the entire search space.<br><br>
The defining characteristic of random search is how the descent direction is determined at each iteration. In this algorithm, a given number of random directions are sampled from the current point, and the point with the lowest evaluation is chosen as the next point to move to. This process is repeated until the stopping criterion is met (e.g. the function value reaches a certain threshold or the number of iterations has been exceeded).
<br><br>
It is important to note that since the algorithm only explores the neighborhood of the current point, it may get stuck in local minima and not find the global minimum. It can be useful in cases when the function is relatively simple or well-behaved, or when the computational cost of evaluating the function is low.
<br><br>
The number of random directions sampled at each iteration is called the population size and is denoted by P. The parameter P can be adjusted to trade-off between exploration and exploitation. A large P value increases the chances of finding a better point, but it also increases the computational cost. A small P value reduces the computational cost but also reduces the chances of finding a better point.

Steps :<br>
- Let the function we want to minimize be denoted as f(x) where x is a vector of parameters.<br><br>
- At each iteration k, we sample a given number of random directions d from the current point $x_k$. The number of directions sampled is denoted by P. The directions are sampled from a probability distribution, usually a normal distribution centered at $ x_k.$<br><br>
- For each sampled direction d_i, we evaluate the function at the point $x_k + d_i$ and obtain the function value $f(x_k + d_i)$.<br><br>
- We choose the direction d_j that gives us the smallest evaluation of the function, so long as it is lower on the function than our current point. In other words, we choose $d_j$ such that $j = argmin{f(x_k + d_i)}$, i = 1,2,..,P
<br><br>
- we update the parameters with the direction that gives us the smallest evaluation of the function: $x_{k+1} = x_k + d_j$<br><br>

It can be summarized in this mathematical form :
<br><br>
$$
d_i ~ N(x_k, \sigma)
$$
where N is the normal distribution 
$$
x_k+1 = x_k + d_j 
$$
where $j = argmin{f(x_k + d_i)}$, i = 1,2,..,P

## 2.4.1 How Does Adam Work?
Adam is different to classical stochastic gradient descent

Stochastic gradient descent maintains a single learning rate (termed alpha) for all weight updates and the learning rate does not change during training.
.

A learning rate is maintained for each network weight (parameter) and separately adapted as learning unfolds.

The method computes individual adaptive learning rates for different parameters from estimates of first and second moments of the gradients.

The authors describe Adam as combining the advantages of two other extensions of stochastic gradient descent. Specifically:

### 2.4.1.1 Adaptive Gradient Algorithm (AdaGrad)
is an extension of the gradient descent optimization algorithm and uses adaptive learning rates to improve the optimization process. Specifically, it adjusts the step size in each dimension of the gradient descent algorithm by keeping track of the sum of the squares of the gradients and using that information to adapt the gradient updates in different directions.


It is clear from the update rule that history of the gradient is accumulated in v. The smaller the gradient accumulated, the smaller the v value will be, leading to a bigger learning rate (because v divides η).

### 2.4.1.2 Root Mean Square Propagation (RMSProp)
Root mean square prop or RMSprop is an adaptive learning algorithm that tries to improve AdaGrad. Instead of taking the cumulative sum of squared gradients like in AdaGrad, it takes the ‘exponential moving average’.<br>
In RMS prop, each update is done according to the equations described below. This update is done separately for each parameter.
<br><br><br>
**sum_of_gradient_squared = previous_sum_of_gradient_squared * decay_rate+ gradient² * (1- decay_rate)**
<br><br><br>
**delta = -learning_rate * gradient / sqrt(sum_of_gradient_squared)**

<br><br>
**$\epsilon$ = a small +ve constant to avoid 'division by 0' error when (vt -> 0). (10-8)** <br>

 ## 2.4.2 Adam — Adaptive Moment Estimation 
Adam Optimizer inherits the strengths or the positive attributes of the above two methods and builds upon them to give a more optimized gradient descent.<br>
Does everything that RMSProp does to solve the denominator decay problem of AdaGrad. In addition to that, use a cumulative history of gradients.

Here, we control the rate of gradient descent in such a way that there is minimum oscillation when it reaches the global minimum while taking big enough steps (step-size) so as to pass the local minima hurdles along the way. Hence, combining the features of the above methods to reach the global minimum efficiently. 

### 2.4.2.1 Mathematical Aspect of Adam Optimizer 
**sum_of_gradient = previous_sum_of_gradient * beta1 + gradient * (1 - beta1) [Momentum]
<br>
sum_of_gradient_squared = previous_sum_of_gradient_squared * beta2 + gradient² * (1- beta2) [RMSProp]
<br>
delta = -learning_rate * sum_of_gradient / sqrt(sum_of_gradient_squared)
<br>
theta += delta**
<br><br>
Taking the formulas used in the above two methods, we get

**$\epsilon$ = a small +ve constant to avoid 'division by 0' error when (vt -> 0). (10-8)** <br>
**β1 & β2 = decay rates of average of gradients in the above two methods. (β1 = 0.9 & β2 = 0.999)**

Here, we compute the exponential average of the gradient as well as the squares of the gradient for each parameters. To decide our learning step, we multiply our learning rate by **average of the gradient** and divide it by **the root mean square** of the exponential average of square of gradients. <br>
The hyperparameter beta1 is generally kept around 0.9 while beta_2 is kept at 0.99. Epsilon is chosen to be 1e-10 generally.

### 2.4.2.2 Implementation of Adam

```python
import numpy as np
class AdamOptim():
    def __init__(self, eta=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.m_dw, self.v_dw = 0, 0
        self.m_db, self.v_db = 0, 0
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.eta = eta
    def update(self, t, w, b, dw, db):
        ## dw, db are from current minibatch
        ## momentum beta 1
        # *** weights *** #
        self.m_dw = self.beta1*self.m_dw + (1-self.beta1)*dw
        # *** biases *** #
        self.m_db = self.beta1*self.m_db + (1-self.beta1)*db

        ## rms beta 2
        # *** weights *** #
        self.v_dw = self.beta2*self.v_dw + (1-self.beta2)*(dw**2)
        # *** biases *** #
        self.v_db = self.beta2*self.v_db + (1-self.beta2)*(db)

        ## bias correction
        m_dw_corr = self.m_dw/(1-self.beta1**t)
        m_db_corr = self.m_db/(1-self.beta1**t)
        v_dw_corr = self.v_dw/(1-self.beta2**t)
        v_db_corr = self.v_db/(1-self.beta2**t)

        ## update weights and biases
        w = w - self.eta*(m_dw_corr/(np.sqrt(v_dw_corr)+self.epsilon))
        b = b - self.eta*(m_db_corr/(np.sqrt(v_db_corr)+self.epsilon))
        return w, b
```

**Testing the implementation** <br>
To test our implementation, we will first need to define a loss function and its respective gradient function. A gradient function can be obtained by simply taking the derivative of the loss function.

```python
def loss_function(m):
    return m**2-2*m+1
## take derivative
def grad_function(m):
    return 2*m-2
def check_convergence(w0, w1):
    return (w0 == w1)
```

Finally, we can iteratively update the weights and biases using our constructed Adam optimizer and see if they converge:

```python
w_0 = 0
b_0 = 0
adam = AdamOptim()
t = 1 
converged = False

while not converged:
    dw = grad_function(w_0)
    db = grad_function(b_0)
    w_0_old = w_0
    w_0, b_0 = adam.update(t,w=w_0, b=b_0, dw=dw, db=db)
    if check_convergence(w_0, w_0_old):
        print('converged after '+str(t)+' iterations')
        break
    else:
        print('iteration '+str(t)+': weight='+str(w_0))
        t+=1
```

Before understanding the Genetic algorithm, let's first understand basic terminologies to better understand this algorithm:

### 2.5.1.2 How Genetic Algorithm Work?

**Initialization:** The process of a genetic algorithm starts by generating the set of individuals, which is called population. Here each individual is the solution for the given problem. ### 2.5.1.4 Implementation: random string using Genetic Algorithm

Given a target string, the goal is to produce target string starting from a random string of the same length. In the following implementation, following analogies are made

* Characters A-Z, a-z, 0-9, and other special symbols are considered as genes
* A string generated by these characters is considered as chromosome/solution/Individual

>**Fitness score** is the number of characters which differ from characters in target string at a particular index. So individual having lower fitness value is given more preference.  

```python
import random
```

```python
# Number of individuals in each generation
POPULATION_SIZE = 100

# Valid genes
GENES = '''abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOP
QRSTUVWXYZ 1234567890, .-;:_!"#%&/()=?@${[]}'''

# Target string to be generated
TARGET = "I love Mathematics"
```

```python
class Individual(object):
    
    #Class representing individual in population
    
    def __init__(self, chromosome):
        self.chromosome = chromosome
        self.fitness = self.cal_fitness()
        
    @classmethod
    def mutated_genes(self):
        #create random genes for mutation
        global GENES
        gene = random.choice(GENES)
        return gene

    @classmethod
    def create_gnome(self):
        #create chromosome or string of genes
        global TARGET
        gnome_len = len(TARGET)
        return [self.mutated_genes() for _ in range(gnome_len)]

    def mate(self, par2):
        #Perform mating and produce new offspring

        # chromosome for offspring
        child_chromosome = []
        for gp1, gp2 in zip(self.chromosome, par2.chromosome):	

            # random probability
            prob = random.random()

            # if prob is less than 0.45, insert gene
            # from parent 1
            if prob < 0.45:
                child_chromosome.append(gp1)

            # if prob is between 0.45 and 0.90, insert
            # gene from parent 2
            elif prob < 0.90:
                child_chromosome.append(gp2)

            # otherwise insert random gene(mutate),
            # for maintaining diversity
            else:
                child_chromosome.append(self.mutated_genes())

        # create new Individual(offspring) using
        # generated chromosome for offspring
        return Individual(child_chromosome)

    def cal_fitness(self):
        #Calculate fitness score, it is the number of characters in string which differ from target string.
        global TARGET
        fitness = 0
        for gs, gt in zip(self.chromosome, TARGET):
            if gs != gt: fitness+= 1
        return fitness
```

```python
global POPULATION_SIZE
  
#current generation
generation = 1
  
found = False
population = []
  
# create initial population
for _ in range(POPULATION_SIZE):
    gnome = Individual.create_gnome()
    population.append(Individual(gnome))
  
while not found:
    
    # sort the population in increasing order of fitness score
    population = sorted(population, key = lambda x:x.fitness)
  
    # if the individual having lowest fitness score ie. 
    # 0 then we know that we have reached to the target
    # and break the loop
    if population[0].fitness <= 0:
        found = True
        break
  
    # Otherwise generate new offsprings for new generation
    new_generation = []
  
    # Perform Elitism, that mean 10% of fittest population
    # goes to the next generation
    s = int((10*POPULATION_SIZE)/100)
    new_generation.extend(population[:s])
  
    # From 50% of fittest population, Individuals 
    # will mate to produce offspring
    s = int((90*POPULATION_SIZE)/100)
    for _ in range(s):
        parent1 = random.choice(population[:50])
        parent2 = random.choice(population[:50])
        child = parent1.mate(parent2)
        new_generation.append(child)
  
    population = new_generation
  
    print("Generation: {}\tString: {}\tFitness: {}".\
          format(generation,
          "".join(population[0].chromosome),
          population[0].fitness))
  
    generation += 1
  
      
print("Generation: {}\tString: {}\tFitness: {}".\
      format(generation,
      "".join(population[0].chromosome),
      population[0].fitness))
```

>**Note:** Every-time algorithm start with random strings, so output may differ

### 2.5.2.1 Bees in nature

A colony of honey bees can extend itself over
long distances (more than 10 km) and in multiple
directions simultaneously to exploit a large number
of food sources. A colony prospers by
deploying its foragers to good fields. In principle,
flower patches with plentiful amounts of nectar or
pollen that can be collected with less effort should be
visited by more bees, whereas patches with less
nectar or pollen should receive fewer bees.

The foraging process begins in a colony by scout
bees being sent to search for promising flower
patches. Scout bees move randomly from one patch
to another. During the harvesting season, a colony
continues its exploration, keeping a percentage of the
population as scout bees.

When they return to the hive, those scout bees
that found a patch which is rated above a certain
quality threshold (measured as a combination of
some constituents, such as sugar content) deposit
their nectar or pollen and go to the “dance floor” to
perform a dance known as the “waggle dance”.

This mysterious dance is essential for colony
communication, and contains three pieces of
information regarding a flower patch: the direction in
which it will be found, its distance from the hive and
its quality rating (or fitness). This information
helps the colony to send its bees to flower patches
precisely, without using guides or maps. Each
individual’s knowledge of the outside environment is
gleaned solely from the waggle dance. This dance
enables the colony to evaluate the relative merit of
different patches according to both the quality of the
food they provide and the amount of energy needed
to harvest it. After waggle dancing on the dance
floor, the dancer (i.e. the scout bee) goes back to the
flower patch with follower bees that were waiting
inside the hive. More follower bees are sent to more
promising patches. This allows the colony to gather
food quickly and efficiently.

While harvesting from a patch, the bees monitor
its food level. This is necessary to decide upon the
next waggle dance when they return to the hive.
If the patch is still good enough as a food source,
then it will be advertised in the waggle dance and
more bees will be recruited to that source.

### 5.2.2 Proposed Bees Algorithm

As mentioned , the Bees Algorithm is an
optimisation algorithm inspired by the natural
foraging behaviour of honey bees to find the optimal
solution. The algorithm
requires a number of parameters to be set, namely:
number of scout bees (n), number of sites selected
out of n visited sites (m), number of best sites out of
m selected sites (e), number of bees recruited for best
e sites (nep), number of bees recruited for the other
(m-e) selected sites (nsp), initial size of patches (ngh)
which includes site and its neighbourhood and
stopping criterion. The algorithm starts with the n
scout bees being placed randomly in the search space.
The fitnesses of the sites visited by the scout bees are
evaluated in step 2.

>1. Initialise population with random solutions.
>2. Evaluate fitness of the population.
>3. While (stopping criterion not met)//Forming new population.
>4. Select sites for neighbourhood search.
>5. Recruit bees for selected sites (more bees
for best e sites) and evaluate fitnesses.
>6. Select the fittest bee from each patch.
>7. Assign remaining bees to search randomly
and evaluate their fitnesses.
>8. End While.

In step 4, bees that have the highest fitnesses are
chosen as “selected bees” and sites visited by them
are chosen for neighbourhood search. Then, in steps
5 and 6, the algorithm conducts searches in the
neighbourhood of the selected sites, assigning more
bees to search near to the best e sites. The bees can
be chosen directly according to the fitnesses
associated with the sites they are visiting.
Alternatively, the fitness values are used to
determine the probability of the bees being selected.
Searches in the neighbourhood of the best e sites
which represent more promising solutions are made
more detailed by recruiting more bees to follow them
than the other selected bees. Together with scouting,
this differential recruitment is a key operation of the
Bees Algorithm.

However, in step 6, for each patch only the bee
with the highest fitness will be selected to form the
next bee population. In nature, there is no such a
restriction. This restriction is introduced here to
reduce the number of points to be explored. In step 7,
the remaining bees in the population are assigned
randomly around the search space scouting for new
potential solutions. These steps are repeated until a
stopping criterion is met. At the end of each iteration,
the colony will have two parts to its new population
– representatives from each selected patch and other
scout bees assigned to conduct random searches. 

### 2.5.2.3 Implementation: Optimize the Radiation pattern of an antenna array

The code starts by defining the `objective_function`, which takes an array of complex numbers (representing the weights of the antenna elements) as input and returns the SIR. The SIR is calculated by taking the ratio of the power of the desired signal to the power of the undesired signals. The objective function is then maximized by minimizing the negative SIR.

The algorithm then iterates for a specified number of iterations, during each iteration, the bees perform a "waggle dance" to generate new solutions and select the best one, and a small probability of "scout bee" phase, where a bee is replaced with a new randomly generated solution. The best solution found is then returned.

```python
def objective_function(x):
    # Define the objective function to be optimized
    # x is an array of complex numbers representing the weights of the antenna elements
    # The function returns the signal-to-interference ratio (SIR)
    signal = np.abs(np.dot(x, desired_signal))**2
    interference = np.abs(np.dot(x, undesired_signals))**2
    SIR = signal / interference
    return -SIR

def generate_random_solution():
    # Generate a random solution
    # The solution is an array of complex numbers representing the weights of the antenna elements
    x = np.random.randn(n_antennas) + 1j * np.random.randn(n_antennas)
    return x

def Bee_Algorithm(objective_function, n_bees, n_iterations):
    # Initialize the algorithm
    solutions = [generate_random_solution() for _ in range(n_bees)]
    best_solution = min(solutions, key=objective_function)
    for _ in range(n_iterations):
        for i in range(n_bees):
            # Perform the waggle dance
            new_solution = solutions[i] + (np.random.randn(n_antennas) + 1j * np.random.randn(n_antennas)) * step_size
            if objective_function(new_solution) < objective_function(solutions[i]):
                solutions[i] = new_solution
            # Perform the scout bee phase
            if random.random() < 0.1:
                solutions[i] = generate_random_solution()
        # Update the best solution
        best_solution = min(solutions, key=objective_function)
    return best_solution
```

```python
# Define the parameters of the problem
n_antennas = 64 # Number of antenna elements
desired_signal = np.exp(1j * np.random.rand(n_antennas) * 2 * np.pi) # The desired signal
undesired_signals = np.random.randn(n_antennas) + 1j * np.random.randn(n_antennas) # The undesired signals
step_size = 0.1 # The step size for the waggle dance
```

Finally, The code defines some parameters of the problem, such as the number of antenna elements, the desired signal, the undesired signals, and the step size, and it runs the Bee Algorithm by calling the `Bee_Algorithm` function and passing the appropriate arguments. The result of the optimization, which is the best set of complex weights for the antenna elements, is then printed.

```python
# Run the Bee Algorithm
result = Bee_Algorithm(objective_function, n_bees=20, n_iterations=100)

# Print the result
print(result)
```

## 2.5.3 Ant colony Algorithm

### 2.5.3.1 What is a Ant Colony optimization Algorithm?

Ant colony optimization(ACO) was first introduced by Marco Dorigo in the 90s in his Ph.D. thesis. This algorithm is introduced based on the foraging behavior of an ant for seeking a path between their colony and source food. Initially, it was used to solve the well-known traveling salesman problem. Later, it is used for solving different hard optimization problems.

Ants are social insects. They live in colonies. The behavior of the ants is controlled by the goal of searching for food. While searching, ants roaming around their colonies. An ant repeatedly hops from one place to another to find the food. While moving, it deposits an organic compound called pheromone on the ground. Ants communicate with each other via pheromone trails. When an ant finds some amount of food it carries as much as it can carry. When returning it deposits pheromone on the paths based on the quantity and quality of the food. Ant can smell pheromone. So, other ants can smell that and follow that path. The higher the pheromone level has a higher probability of choosing that path and the more ants follow the path, the amount of pheromone will also increase on that path.

### 2.5.3.4 Implementation: Traveling Salesman Problem (TSP)

This code is an example of how the Ant Colony Optimization (ACO) algorithm can be implemented for a given Traveling Salesman Problem (TSP).

```python
import numpy as np
```

The first section of the code defines the TSP instance, which includes the number of cities (n) and the distances between the cities represented in a 2D matrix.

```python
# TSP instance
n = 5  # number of cities
distances = np.array([[0, 2, 9, 10, 11],
                     [1, 0, 6, 4, 3],
                     [15, 7, 0, 8, 10],
                     [6, 14, 9, 0, 2],
                     [8, 5, 12, 6, 0]])  # distances between the cities
```

This section of the code sets the parameters for the ACO algorithm, including the number of ants (m), the pheromone influence (alpha), the distance influence (beta), the pheromone evaporation rate (rho), the pheromone deposit constant (Q), and the maximum number of iterations.

```python
# ACO parameters
m = 3  # number of ants
alpha = 1  # pheromone influence
beta = 2  # distance influence
rho = 0.1  # pheromone evaporation rate
Q = 1  # pheromone deposit constant
max_iterations = 100  # maximum number of iterations
```

The algorithm starts by initializing the pheromone levels on the edges with a small value, and initialize the best solution with infinite cost.

The main loop of the algorithm runs for the maximum number of iterations defined, on each iteration the following operations are performed:

* Initialize solutions for each ant
* each ant starts from a random city, and construct a solution by repeatedly choosing the next city based on a probability distribution that is influenced by the current pheromone levels on the edges and the distances between the cities.
* update the best solution if the current solution is better.
* update the pheromone levels on the edges based on the quality of the solutions constructed by the ants, with better solutions leading to stronger pheromone trails.

The loop continues till maximum number of iteration defined or till a satisfactory solution is found.

```python
# Initialize pheromone levels
tau = np.ones((n, n)) / n

# Initialize the best solution
best_solution = None
best_cost = float('inf')

for iteration in range(max_iterations):
    # Initialize solutions for each ant
    solutions = []
    costs = []
    for i in range(m):
        # Start each ant in a random city
        current_city = np.random.randint(n)
        unvisited_cities = set(range(n)) - {current_city}
        solution = [current_city]
        cost = 0
        while unvisited_cities:
            # Choose next city based on probability distribution
            next_city_probs = np.array([np.power(tau[current_city, j], alpha) *
                                       np.power(1 / distances[current_city, j], beta)
                                       for j in unvisited_cities])
            next_city_probs = next_city_probs / sum(next_city_probs)
            next_city = np.random.choice(list(unvisited_cities), p=next_city_probs)
            
            # Update solution and cost
            solution.append(next_city)
            cost += distances[current_city, next_city]
            current_city = next_city
            unvisited_cities.remove(next_city)
        
        # Update best solution
        if cost < best_cost:
            best_solution = solution
            best_cost = cost
        
        # Append solution and cost
        solutions.append(solution)
        costs.append(cost)
    
    # Update pheromone levels
    for i in range(n):
        for j in range(n):
            tau[i, j] *= (1 - rho)
            for k in range(m):
                if i in solutions[k] and j in solutions[k]:
                    tau[i, j] += Q / costs[k]
```
